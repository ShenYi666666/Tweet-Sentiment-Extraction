```python
train = pd.read_csv('../input/tweet-sentiment-extraction/tweet_dataset.csv')
```

```python
def get_pos(x, y):
    return x.find(y)
```

```python
train['start_pos'] = train.apply(lambda x: get_pos(x['text'], x['selected_text']), axis=1)
train['end_pos'] = train['start_pos']+train['selected_text'].str.len()
```

```python
def broken_start(x, y):
    
    if y>0 and x[y-1].isalpha():
        return True
    return False

def broken_end(x, y):
    if y<len(x) and x[y]!=' ':
        return True
    return False
```

```python
train['broken_start'] = train.apply(lambda x: broken_start(x['text'], x['start_pos']), axis=1)
train['broken_end'] = train.apply(lambda x: broken_end(x['text'], x['end_pos']), axis=1)
train['broken'] = (train['broken_start'] | train['broken_end'])
```

```python
def find_last_char(x, y):
    if y>1 and x[y-1]==' ':
        return x[y-2]
    return np.nan
```

```python
train['prev_word_last_char'] = train.apply(lambda x: find_last_char(x['text'], x['start_pos']), axis=1)
```

```python
def get_prev_word(x, y):
    if y>1 and x[y-1]==' ':
        s = y-2
        while s>=0 and x[s]!=' ':
            s = s-1
        return x[s+1:y-1]
    if y>1 and x[y-1]!=' ':
        s = y-1
        while s>=0 and x[s]!=' ':
            s = s-1
        e = y
        while e<len(x) and x[e].isalpha():
            e = e+1
        return x[s+1:e]
    return np.nan
```

```python
train['prev_word'] = train.apply(lambda x: get_prev_word(x['text'], x['start_pos']), axis=1)
```
